```python
import pandas as pd

# Load data for all teams
boston_regular = pd.read_csv('Boston_Celtics_Regular_Season_Stats_2023.csv')
boston_playoff = pd.read_csv('Boston_Celtics_Playoff_Stats_2023.csv')
dallas_regular = pd.read_csv('Dallas_Mavericks_Regular_Season_Stats_2023.csv')
dallas_playoff = pd.read_csv('Dallas_Mavericks_Playoff_Stats_2023.csv')
minnesota_regular = pd.read_csv('Minnesota_Timberwolves_Regular_Season_Stats_2023.csv')
minnesota_playoff = pd.read_csv('Minnesota_Timberwolves_Playoff_Stats_2023.csv')
indiana_regular = pd.read_csv('Indiana_Pacers_Regular_Season_Stats_2023.csv')
indiana_playoff = pd.read_csv('Indiana_Pacers_Playoff_Stats_2023.csv')
newyork_regular = pd.read_csv('New_York_Knicks_Regular_Season_Stats_2023.csv')
newyork_playoff = pd.read_csv('New_York_Knicks_Playoff_Stats_2023.csv')
lakers_regular = pd.read_csv('Los_Angeles_Lakers_Regular_Season_Stats_2023.csv')
lakers_playoff = pd.read_csv('Los_Angeles_Lakers_Playoff_Stats_2023.csv')

# Print the first few rows of each DataFrame to verify successful loading
print("Boston Celtics Regular Season Data:")
print(boston_regular.head())

print("\nBoston Celtics Playoff Data:")
print(boston_playoff.head())

print("\nDallas Mavericks Regular Season Data:")
print(dallas_regular.head())

print("\nDallas Mavericks Playoff Data:")
print(dallas_playoff.head())

print("\nMinnesota Timberwolves Regular Season Data:")
print(minnesota_regular.head())

print("\nMinnesota Timberwolves Playoff Data:")
print(minnesota_playoff.head())

print("\nIndiana Pacers Regular Season Data:")
print(indiana_regular.head())

print("\nIndiana Pacers Playoff Data:")
print(indiana_playoff.head())

print("\nNew York Knicks Regular Season Data:")
print(newyork_regular.head())

print("\nNew York Knicks Playoff Data:")
print(newyork_playoff.head())

print("\nLos Angeles Lakers Regular Season Data:")
print(lakers_regular.head())

print("\nLos Angeles Lakers Playoff Data:")
print(lakers_playoff.head())
```

```python
# List of DataFrames to clean
dataframes = [boston_regular, boston_playoff, dallas_regular, dallas_playoff,
              minnesota_regular, minnesota_playoff, indiana_regular, indiana_playoff,
              newyork_regular, newyork_playoff, lakers_regular, lakers_playoff]

# Function to clean each DataFrame
def clean_data(df):
    # Drop unnecessary columns (like 'Rk', 'Unnamed: 17', 'Unnamed: 22')
    df = df.drop(columns=['Rk', 'Unnamed: 17', 'Unnamed: 22'])

    # Fill missing values
    df = df.fillna(0)

    # Ensure correct data types
    for col in df.columns:
        if col not in ['Player']:
            df[col] = pd.to_numeric(df[col], errors='coerce')

    return df

# Apply the cleaning function to each DataFrame
boston_regular = clean_data(boston_regular)
boston_playoff = clean_data(boston_playoff)
dallas_regular = clean_data(dallas_regular)
dallas_playoff = clean_data(dallas_playoff)
minnesota_regular = clean_data(minnesota_regular)
minnesota_playoff = clean_data(minnesota_playoff)
indiana_regular = clean_data(indiana_regular)
indiana_playoff = clean_data(indiana_playoff)
newyork_regular = clean_data(newyork_regular)
newyork_playoff = clean_data(newyork_playoff)
lakers_regular = clean_data(lakers_regular)
lakers_playoff = clean_data(lakers_playoff)

# Verify cleaning by checking the first few rows of each DataFrame
print("Cleaned Boston Celtics Regular Season Data:")
print(boston_regular.head())

print("\nCleaned Boston Celtics Playoff Data:")
print(boston_playoff.head())

print("\nCleaned Dallas Mavericks Regular Season Data:")
print(dallas_regular.head())

print("\nCleaned Dallas Mavericks Playoff Data:")
print(dallas_playoff.head())

print("\nCleaned Minnesota Timberwolves Regular Season Data:")
print(minnesota_regular.head())

print("\nCleaned Minnesota Timberwolves Playoff Data:")
print(minnesota_playoff.head())

print("\nCleaned Indiana Pacers Regular Season Data:")
print(indiana_regular.head())

print("\nCleaned Indiana Pacers Playoff Data:")
print(indiana_playoff.head())

print("\nCleaned New York Knicks Regular Season Data:")
print(newyork_regular.head())

print("\nCleaned New York Knicks Playoff Data:")
print(newyork_playoff.head())

print("\nCleaned Los Angeles Lakers Regular Season Data:")
print(lakers_regular.head())

print("\nCleaned Los Angeles Lakers Playoff Data:")
print(lakers_playoff.head())
```

```python
# Function to calculate average metrics
def calculate_average_metrics(df, team_name, season_type):
    metrics = {
        'PER': df['PER'].mean(),
        'TS%': df['TS%'].mean(),
        'TRB%': df['TRB%'].mean(),
        'AST%': df['AST%'].mean(),
        'USG%': df['USG%'].mean()
    }
    return pd.DataFrame(metrics, index=[f"{team_name}_{season_type}"])

# Calculate metrics for each team's regular season and playoff data
teams = ['Boston Celtics', 'Dallas Mavericks', 'Minnesota Timberwolves',
         'Indiana Pacers', 'New York Knicks', 'Los Angeles Lakers']

season_types = ['Regular', 'Playoff']

dfs = {
    'Boston Celtics_Regular': boston_regular,
    'Boston Celtics_Playoff': boston_playoff,
    'Dallas Mavericks_Regular': dallas_regular,
    'Dallas Mavericks_Playoff': dallas_playoff,
    'Minnesota Timberwolves_Regular': minnesota_regular,
    'Minnesota Timberwolves_Playoff': minnesota_playoff,
    'Indiana Pacers_Regular': indiana_regular,
    'Indiana Pacers_Playoff': indiana_playoff,
    'New York Knicks_Regular': newyork_regular,
    'New York Knicks_Playoff': newyork_playoff,
    'Los Angeles Lakers_Regular': lakers_regular,
    'Los Angeles Lakers_Playoff': lakers_playoff
}

average_metrics = []

for team in teams:
    for season_type in season_types:
        df_name = f"{team}_{season_type}"
        average_metrics.append(calculate_average_metrics(dfs[df_name], team, season_type))

# Concatenate all average metrics into a single DataFrame
average_metrics_df = pd.concat(average_metrics)

# Reset index for better readability
average_metrics_df.reset_index(inplace=True)
average_metrics_df.rename(columns={'index': 'Team_Season'}, inplace=True)

# Display the average metrics
print("Average Metrics for Regular Season and Playoffs:")
print(average_metrics_df)
```

```python
# Identifying top contributors for Lakers regular season
top_contributors_regular = lakers_regular_metrics.sort_values(by='PER', ascending=False).head(5)
print("Top Contributors for Lakers Regular Season:")
print(top_contributors_regular)

# Identifying top contributors for Lakers playoffs
top_contributors_playoff = lakers_playoff_metrics.sort_values(by='PER', ascending=False).head(5)
print("\nTop Contributors for Lakers Playoffs:")
print(top_contributors_playoff)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Function to plot key metrics
def plot_key_metrics(data, title):
    fig, axs = plt.subplots(3, 1, figsize=(10, 15))

    sns.barplot(x='Player', y='PER', data=data, ax=axs[0])
    axs[0].set_title('PER')
    axs[0].set_ylabel('PER')

    sns.barplot(x='Player', y='TS%', data=data, ax=axs[1])
    axs[1].set_title('TS%')
    axs[1].set_ylabel('TS%')

    sns.barplot(x='Player', y='TRB%', data=data, ax=axs[2])
    axs[2].set_title('TRB%')
    axs[2].set_ylabel('TRB%')

    plt.suptitle(title)
    plt.show()

# Plotting key metrics for top contributors in the regular season
plot_key_metrics(top_contributors_regular, 'Lakers Regular Season Top Contributors')

# Plotting key metrics for top contributors in the playoffs
plot_key_metrics(top_contributors_playoff, 'Lakers Playoff Top Contributors')
```

```python
# Function to calculate key metrics and identify top contributors
def calculate_key_metrics(df):
    key_metrics = df[['Player', 'PER', 'TS%', 'TRB%', 'AST%', 'USG%']]
    return key_metrics.sort_values(by='PER', ascending=False)

# Calculate key metrics for Lakers regular season and playoffs
lakers_regular_metrics = calculate_key_metrics(lakers_regular)
lakers_playoff_metrics = calculate_key_metrics(lakers_playoff)

# Display the top contributors for Lakers regular season and playoffs
print("Top Contributors for Lakers Regular Season:")
print(lakers_regular_metrics.head(5))

print("\nTop Contributors for Lakers Playoffs:")
print(lakers_playoff_metrics.head(5))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing the top contributors
def plot_top_contributors(metrics, title):
    top_metrics = metrics.head(5)
    fig, axes = plt.subplots(3, 1, figsize=(10, 15))
    fig.suptitle(title, fontsize=16)

    sns.barplot(x='Player', y='PER', data=top_metrics, ax=axes[0])
    axes[0].set_title('PER')

    sns.barplot(x='Player', y='TS%', data=top_metrics, ax=axes[1])
    axes[1].set_title('TS%')

    sns.barplot(x='Player', y='TRB%', data=top_metrics, ax=axes[2])
    axes[2].set_title('TRB%')

    plt.tight_layout()
    plt.show()

plot_top_contributors(lakers_regular_metrics, 'Lakers Regular Season Top Contributors')
plot_top_contributors(lakers_playoff_metrics, 'Lakers Playoff Top Contributors')
```

```python
# Function to compare regular season and playoff performances
def compare_performances(regular_metrics, playoff_metrics):
    comparison = pd.merge(regular_metrics, playoff_metrics, on='Player', suffixes=('_Regular', '_Playoff'))
    comparison['PER_Change'] = comparison['PER_Playoff'] - comparison['PER_Regular']
    comparison['TS%_Change'] = comparison['TS%_Playoff'] - comparison['TS%_Regular']
    comparison['TRB%_Change'] = comparison['TRB%_Playoff'] - comparison['TRB%_Regular']
    comparison['AST%_Change'] = comparison['AST%_Playoff'] - comparison['AST%_Regular']
    comparison['USG%_Change'] = comparison['USG%_Playoff'] - comparison['USG%_Regular']
    return comparison

# Compare Lakers regular season and playoff performances
lakers_comparison = compare_performances(lakers_regular_metrics, lakers_playoff_metrics)

# Display the comparison
print("Lakers Regular Season vs Playoff Performance Comparison:")
print(lakers_comparison)
```

```python
# Contextualizing Contributions
def contextualize_contributions(comparison):
    # Identifying key areas for improvement based on player metrics
    rebounding_needs = comparison[comparison['TRB%_Change'] < 0]
    scoring_needs = comparison[comparison['TS%_Change'] < 0]
    playmaking_needs = comparison[comparison['AST%_Change'] < 0]

    print("\nPlayers with Decreased Rebounding in Playoffs:")
    print(rebounding_needs[['Player', 'TRB%_Change']])

    print("\nPlayers with Decreased Scoring Efficiency in Playoffs:")
    print(scoring_needs[['Player', 'TS%_Change']])

    print("\nPlayers with Decreased Playmaking in Playoffs:")
    print(playmaking_needs[['Player', 'AST%_Change']])

contextualize_contributions(lakers_comparison)
```
